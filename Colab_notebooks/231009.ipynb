```python
import tensorflow as tf
import numpy as np

a = np.ones((2, 3), dtype=np.float32) # (2.3)의 numpy 배열 생성, 1로 초기화
b = tf.ones([3, 1], dtype=tf.float32) # [3,1] 리스트를 1로 초기화 하여 상수형 텐서 생성

print(tf.matmul(a, b)) # numpy 배열을 텐서로 자동형변환 후 곱 연산 후 텐서로 출력
```

```python
import tensorflow as tf
var = tf.Variable(tf.ones((2, 3))) # 배열 상수형 텐서를 변수형 텐서로 변환
print(var)*
var1 = var + 2 # var은 tf.Tensor로 자동 형변환이 일어나 상수 2와 연산되어 텐서 var1에 저장된다.

print(var1)

```

```python
import tensorflow as tf

a = tf.range(10, delta=2, dtype=tf.int32) # 0부터 9까지 2씩 건너뛴 벡터 텐서
print(a)
b = 3 * tf.ones(5, dtype=tf.int32) # 1로 구성된 벡터 텐서
print(b)

print(tf.add(a,b))
print(a+b)

print(tf.multiply(a,b))
print(a*b)
```

```python
import tensorflow as tf

x = tf.constant([0.9, 2.6, 2.3, 1.5, -4.5])
a = tf.round(x) # 반올림
b = tf.floor(x) # 버림

print(a)
print(b)
print(tf.maximum(a,b)) # 두 벡터를 비교해서 각 요소별 최대값
print(tf.cumsum(a, axis=0)) # 벡터 a의 요소 값들을 행 기준(axis=0) 으로 누적합
print(tf.math.cumprod(b, axis=0))
```
