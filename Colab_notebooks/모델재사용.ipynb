```python
import tensorflow as tf
import numpy as np
import os
from keras.models import Sequential
from keras.layers import Dense

data = np.loadtxt('/content/gdrive/My Drive/Colab_data/data.csv', delimiter=',', skiprows=1, unpack=True, dtype=np.float32)
# data = np.loadtxt('C:/Users/Iris-training.csv', delimiter=',', skiprows=1, unpack=True, dtype=np.float32)

x_data = np.transpose(data[0:2])
y_data = np.transpose(data[2:])

model = Sequential()
model.add(Dense(input_dim=2, units=10, activation='relu'))
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

print("=====================initial weights====================")
for weight in model.weights:
  print(weight)

model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.01)
,loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()

# 체크포인트 경로 설정
checkpoint_path = "/content/gdrive/My Drive/Colab_data/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
# 체크포인트 콜백 만들기, 체크포인트 파일 경로, (중간) 가중치 저장여부

# cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path, save_weights_only=True, verbose=1)

# history = model.fit(x_data, y_data, epochs=100, batch_size=1, callbacks = [cp_callback])

# 모델 가중치 로드
model.load_weights(checkpoint_path)
# 로드된 가중치로 테스트
loss, acc = model.evaluate(x_data, y_data, verbose=2)
print("훈련된 모델의 정확도: {:5.2f}%".format(100*acc))

print("=====================weights============================")
for weight in model.weights:
  print(weight)

print("=====================test results=========================")
print(x_data)
print(model.predict(x_data))
print("\n Accuracy: %.4f" % (model.evaluate(x_data, y_data)[1]))

print("==========================================================")
```
