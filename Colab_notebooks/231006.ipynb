```python
import tensorflow as tf
import numpy as np
import matplotlib as mpl
import matplotlib.pylab as plt

X = [1., 2., 3.]

# tf.Variable: 그래프를 계산하면서 최적화 할 변수
# tf.random_normal: 각 변수들의 초기값을 정규분포 랜덤 값으로 초기화
W = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b = tf.Variable(tf.random.normal([1], dtype=tf.float32))

@tf.function # tf.function
def forward(x1):
  return x1 * W + b # 일차 식
Y = forward(X)

print(Y)
plt.plot(X, Y, 'ro')
plt.plot(X,Y)
plt.show()

plt.title("Plot")
plt.plot([1, 4, 9, 16], 'ro')
plt.show()

plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x_data 정의
x_data = np.arange(-2.0, 2.0, 0.1)

# y 데이터 생성: 각각의 x에 대한 1차, 2차, 3차 함수 값 계산
y_linear = x_data # y = x (1차 함수)
y_quadratic = x_data**2 # y = x^2 (2차 함수)
y_cubic = x_data**3 # y = x^3 (3차 함수)

# 그래프 생성
plt.figure(figsize=(10,6))

# 각각의 데이터에 대한 그래프 그리기
plt.plot(x_data, y_linear, label='linear')
plt.plot(x_data, y_linear, 'ro')

plt.plot(x_data, y_quadratic,label='quadratic')
plt.plot(x_data, y_quadratic,'ro')

plt.plot(x_data,y_cubic,label='cubic')
plt.plot(x_data,y_cubic, 'ro')

# 레이블 및 제목 설정
plt.xlabel('x')
plt.ylabel('y')

# 범례 표시 설정
plt.legend()

# 그래프 출력
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# x_data 정의
x_data = np.arange(-2.0, 2.0, 0.1)

# W와 b를 랜덤으로 생성
W_linear = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b_linear = tf.Variable(tf.random.normal([1], dtype=tf.float32))

W_quadratic = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b_quadratic = tf.Variable(tf.random.normal([1], dtype=tf.float32))

# y 데이터 생성: 각각의 x에 대한 1차 및 2차 함수 값 계산
y_linear = W_linear * x_data + b_linear # y = Wx + b (1차 함수)
y_quadratic = W_quadratic * (x_data**2) + b_quadratic # y = Wx^2 + b (2차 함수)

# 그래프 생성
plt.figure(figsize=(10,6))

# 각각의 데이터에 대한 그래프 그리기
plt.plot(x_data, y_linear, label='linear')
plt.plot(x_data, y_linear, 'ro')

plt.plot(x_data, y_quadratic,label='quadratic')
plt.plot(x_data, y_quadratic, 'ro')

# 레이블 및 제목 설정
plt.xlabel('x')
plt.ylabel('y')

# 범례 표시 설정
plt.legend()

# 그래프 출력
plt.show()
```
